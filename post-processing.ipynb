交叉驗證

```python
def REPORT(estimator, X, Y, kf):
    accuracy = []
    
    X=pd.DataFrame(X)
    #kf = KFold(n_splits=kf, shuffle=False)
    #for idx_tr, idx_tes in kf.split(range(int(len(X)))):

    kf = StratifiedKFold(n_splits=kf, random_state=None, shuffle=False)
    for idx_tr, idx_tes in kf.split(X,Y):
        
        # training
        estimator.fit(X.iloc[idx_tr], Y[idx_tr])

        # testing
        pred = estimator.predict(X.iloc[idx_tes])
        label = Y[idx_tes]

        # report
        kp = cohen_kappa_score(pred, label)
        acc = accuracy_score(pred, label)
        print(classification_report(y_pred=pred, y_true=label))
        print("kappa_score",kp)
        accuracy.append(acc)
    return accuracy
```

# Smoothing

```python
proba_0=[]
proba_1=[]

proba = clf1.predict_proba(new_test_X)
for i in range(0,len(proba)):
    p0 = proba[i][0]
    p1 = proba[i][1]
    proba_0.append(p0)
    proba_1.append(p1)
```

```python
mean_proba_0=[]
mean_proba_1=[]

for i in range(0,len(proba_0),fit_time):
    mp0 = np.mean(proba_0[i:i+fit_time])
    mp1 = np.mean(proba_1[i:i+fit_time])
    mean_proba_0.append(mp0)
    mean_proba_1.append(mp1)
```

```python
new_predict_XX=[]
for i in range(len(mean_proba_0)):
    if mean_proba_0[i] > 0.5:
        abc=0
    else:
        abc=1
    new_predict_XX.append(abc)
```

```python
new_predict_XX = np.array(new_predict_XX)

print(classification_report(y_pred = new_predict_XX ,y_true = new_predict_Y, target_names=target_names))
print("kappa_score",cohen_kappa_score(new_predict_XX, new_predict_Y))
```
